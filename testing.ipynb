```python
import pandas as pd
df = pd.read_csv("Phoen.csv")
df.sample(10)
```

```python
df.rename(columns={"Transaction Statement for 8540015748": "Datetime"}, inplace=True)
df.rename(columns={"Unnamed: 1": "Name"}, inplace=True)
df.rename(columns={"Unnamed: 2": "Type"}, inplace=True)
df.rename(columns={"Unnamed: 3": "Amount"}, inplace=True)
df.sample(10)
```

```python
df.info()
```

```python
df = df.dropna()
df.sample(10)
```

```python
df['Amount'] = df['Amount'].str.replace('₹', '')
df.sample(10)
```

```python
df = df[df['Amount'] != 'Amount']
df.sample(10)
```

```python
df = df[df['Type'] != 'Type']
df.sample(10)
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'], format='%b %d, %Y', errors='coerce')
df.sample(10)
```

```python
df = df.dropna(subset=['Datetime'])
df.sample(10)
```

```python
df = df[df['Name'] != 'Paid to']
df.sample(10)
```

```python
df = df[df['Name'] != 'Received from']
df.sample(10)
```

```python
df.info()
```

```python
def extract_name(row):
    if row['Type'] == 'DEBIT':
        return row['Name'].split('Paid to ')[-1]
    elif row['Type'] == 'CREDIT':
        return row['Name'].split('Received from ')[-1]
    else:
        return row['Name']
# Apply the function to create a new column
df["Name"] = df.apply(extract_name, axis=1)
```

```python
df.sample(10)
```

```python
nan_values = df[df.isna().any(axis=1)]

# Print rows with NaN values
print(nan_values)
```

```python
import calendar
df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['Month'] = df['Month'].apply(lambda x: calendar.month_name[int(x)])
df['Date'] = df['Datetime'].dt.day
df['Day_of_Week'] = df['Datetime'].dt.day_name()
```

```python
df.sample(10)
```

```python
df.to_csv('finaldf2.csv')
```

```python
df.dtypes
```

```python
def extract_amount(row):
    if '.' in row['Amount']:
        return row['Amount'].split('.')[0]
    else:
        return row['Amount']
df['Amount'] = df.apply(extract_amount, axis=1)
```

```python
df['Amount'] = df['Amount'].str.replace(',', '').astype(int)
```

# Data Analytics

```python
total_debit = df[df['Type'] == 'DEBIT']['Amount'].sum()
total_credit = df[df['Type'] == 'CREDIT']['Amount'].sum()

print(f"Total Debit: {total_debit}")
print(f"Total Credit: {total_credit}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x=['Debit', 'Credit'], y=[total_debit, total_credit])
plt.xlabel('Transaction Type')
plt.ylabel('Total Amount')
plt.title('Total Debit vs Credit Amount')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Grouping the data by month and summing the amounts
monthly_data = df.groupby('Month')['Amount'].sum()

# Creating the bar chart with rotated x-axis labels
plt.figure(figsize=(10, 6))
plt.bar(monthly_data.index, monthly_data.values)
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.title('Total Amount Spent by Month')
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.show()
```

```python
import matplotlib.pyplot as plt

# Grouping the data by month and summing the amounts
monthly_data = df.groupby('Month')['Amount'].sum()

# Creating a color palette for the bars
colors = plt.cm.tab20c.colors[:len(monthly_data)]

# Creating the bar chart with customizations
plt.figure(figsize=(12, 6))
bars = plt.bar(monthly_data.index, monthly_data.values, color=colors, edgecolor='black')

# Adding grid lines and labels
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel('Month', fontsize=12, fontweight='bold')
plt.ylabel('Total Amount', fontsize=12, fontweight='bold')
plt.title('Total Amount Spent by Month', fontsize=14, fontweight='bold')

# Rotating x-axis labels and setting font style
plt.xticks(rotation=45)
plt.yticks(fontsize=10)

# Adding data labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 20, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
